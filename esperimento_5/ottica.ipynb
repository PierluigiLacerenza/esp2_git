```python
import os
import numpy as np
import pandas as pd
import random as rn
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
def reciproco(v):
    reci=[]
    for i in v:
        reci.append(1/i)
    return reci
```

```python
data=pd.read_excel('ottica.xlsx', header=None)
p=data.iloc[:,0]
q=data.iloc[:,1]

uq=data.iloc[:,2]
p_reciproco=reciproco(p)
q_reciproco=reciproco(q)
```

```python
plt.figure(figsize=(12, 12))
plt.errorbar(p_reciproco, q_reciproco, fmt='o', color='blue', label='Punt', capsize=5)
plt.ylabel('1/p')  
plt.xlabel('1/q')  
plt.title('Linearizzare')  

sns.regplot(x=p_reciproco, y=q_reciproco, ci=None, data=data)
slope, intercept, r_value, p_value, std_err = stats.linregress(p_reciproco, q_reciproco)
plt.legend(loc='upper left')
print("1/f=",intercept)
print(1/intercept)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True) 
plt.show()
```
